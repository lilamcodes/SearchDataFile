```python
import pandas as pd
```

```python
sentences = pd.read_csv("chat_data.csv")
```

```python
sentences.head()
```

```python
# Only keep the conversation column
sentences = sentences[["text"]]
```

## Clustering

```python
import numpy as np
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
```

```python
X = vectorizer.fit_transform(list(sentences["text"])) 
```

```python
X.shape #(Number of sentences, number of distinct words/vocabulary size)
```

```python
from sklearn.cluster import KMeans
true_k = 50
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print ("Cluster %d:" % i)
    for ind in order_centroids[i, :10]:
        print (' %s' % terms[ind])
    print
```

```python
labels = model.fit_predict(X)
```

```python
import pickle
with open("labels.pkl", "wb") as f:
    pickle.dump(labels, f)
```

```python
with open("labels.pkl", "rb") as f:
    labels = pickle.load(f)
```

```python
sample_sentences = []
for cluster in np.unique(labels):
    sentence_for_cluster = sentences.loc[labels == cluster]
    random_sample = sentence_for_cluster.sample(3, replace=True)
    
    for line in random_sample.itertuples():
        sample_sentences.append(line.text)
    sample_sentences.append("----------------")
    
    
    display(random_sample)
    print("----------------")
```

```python
for i in sample_sentences:
    print(i)
    #print("---------------")
```

```python
out_df = pd.DataFrame(sample_sentences)
```
